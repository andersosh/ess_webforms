```python
from openpyxl import Workbook, load_workbook
from openpyxl.worksheet.table import Table
from openpyxl.utils import get_column_letter

wb = Workbook()


destination_filename = "table.xlsx"

ws1 = wb.active
ws1.title = "Table Worksheet"

data = [
    ["Leopard", 9000, 400, 899],
    ["Anmial 2", 235, 2134, 98052],
    ["Animal 3", 25634, 25342, 87643],
    ["Animal 4", 67, 6432, 7543],
]

ws1.append(["Animal", "Ate (lb)", "Weight", "Legs"])

for row in data:
    ws1.append(row)


tableObject = Table(displayName="AnimalsTable",
                    ref="A1:D4")

ws1.add_table(tableObject)


wb.save(destination_filename)
```

```python
from openpyxl import Workbook, load_workbook
from openpyxl.utils import get_column_letter

wb = load_workbook(filename = "new_workbook.xlsx")
destination_filename = "new_workbook_row_col_edit.xlsx"

ws1 = wb["PoopyNuts"]

#ws.insert_what(where,howmany)

ws1.insert_cols(3,4)
ws1.insert_rows(3,5)

ws1.delete_cols(3,4)
ws1.delete_rows(3,5)

#set this way will make zero change. Create and delete same rows. Noice


ws1.merge_cells("C2:E3")
ws1.unmerge_cells("C2:E3")


ws1.merge_cells(start_row=4, start_column=2, end_row=6, end_column=5)
ws1.unmerge_cells(start_row=4, start_column=2, end_row=6, end_column=5)

ws1.row_dimensions.group(2,4,hidden=True)
ws1.column_dimensions.group("B","F",hidden=True)

wb.save(destination_filename)
```

```python
from openpyxl import Workbook
from openpyxl.utils import get_column_letter

wb = Workbook()

ws1 = wb.active
ws1.title = "Employee Timelog"

destination_filename = "Employee_timelog.xlsx"


ws1["A1"] = "Day of the Week"
ws1["B1"] = "Date"
ws1["C1"] = "Start Time"
ws1["D1"] = "End Time"


column_widths = []

for each_column in ws1.columns:
    max_length = 0
    column_name = each_column[0].column

    for each_cell in each_column:
        try:
            
            if len(str(each_cell.value)) > max_length:
                max_length = len(each_cell.value)
                
        except:
            pass


    new_width = max_length
    ws1.column_dimensions.width = new_width


wb.save(destination_filename)
```
