```python
import pandas as pd
import csv
import openpyxl
import datetime
from datetime import datetime
from openpyxl import Workbook, load_workbook
from openpyxl.utils import get_column_letter
from pdfrw import PdfWriter


# save data csv file as xlsx in order to manipulate it
wb = openpyxl.Workbook()
ws = wb.active
with open("Reimbursement Request.csv") as f:
    reader = csv.reader(f, delimiter=",")
    for row in reader:
        ws.append(row)
        
wb.save("Reimbursement Request.xlsx")
```
